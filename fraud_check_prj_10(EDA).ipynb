```python
## import library 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
fc_data =pd.read_csv("D:/Lecture file/25 th lecture 29 march/10. Random Forests/Fraud_check.csv")
fc_data.head()
```

### EDA and Graph

```python
fc_data.info()
```

```python
tax_income =[]
for income in fc_data['Taxable.Income']:
    if income <= 30000 :
        tax_income.append("Risky")
    else:
        tax_income.append("Good")
fc_data['tax_income'] = tax_income
fc_data
```

```python
fc_data_drop = fc_data.drop(columns=['Taxable.Income'])
fc_data_drop
```

```python
fc_data_drop.isnull().sum()
```

#### EDA Analysis on fraud check data by graph

```python
sns.barplot(x='Marital.Status' ,y ='tax_income' ,hue ='Urban',data=fc_data_drop)
plt.title("Tax_income based on  marital status")
plt.savefig('Tax_income based on  marital status.png')
plt.show()
```

```python
sns.barplot(x='Undergrad' ,y ='tax_income' ,hue ='Urban',data=fc_data_drop)
plt.title("Tax_income based on  underaged  status")
plt.savefig('D:/Tax_income based on  undrage status.png')
plt.show()
```

```python
sns.lineplot(data=fc_data_drop.sort_values('tax_income'), x='tax_income', y='Work.Experience', hue='Undergrad')
plt.title(" line plot on Tax_income based on Work.Experience")
plt.savefig('D:/Tax_income based on Work.Experience.png')
```

```python
categories_data = fc_data_drop.select_dtypes(include ='object')
categories_data
```

```python
le = LabelEncoder()
fc_data_enocode = fc_data_drop.copy()

for col in categories_data :
    fc_data_enocode[col] =le.fit_transform(fc_data_enocode[col])
```

```python
plt.figure(figsize=(10, 6))
fc=sns.heatmap(fc_data_enocode.corr(numeric_only=True), annot=True, cmap='crest',linewidth=.5)
plt.title('correalation map on fraud check data')
fc.set(xlabel="", ylabel="")
plt.savefig('D:/correlation plot(c_d).png')
```
